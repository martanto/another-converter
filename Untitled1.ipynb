```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))

i = 1

for label, band in bands:
    ax = plt.subplot(n_bands, 1, i)
    band.plot(ax=ax)
    plt.title(label)
    plt.xlabel("")
    plt.legend(ncol=4)
    i += 1
plt.tight_layout()
plt.show()
```
